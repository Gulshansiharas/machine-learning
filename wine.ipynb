```python
#here we are preparing data
x=d1[['fixed acidity','volatile acidity','alcohol']].values
y=d1['quality'].values
x
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train, x_test, y_train,y_test
```

```python
#Build and train the multiple linear regression model
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#make prediction
y_pred=regressor.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df2=df1.head(10)
df2
```

```python
df2.plot(kind='bar',figsize=(12,22))
plt.show()
```

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print("Mean squared error:",metrics.mean_squared_error(y_test,y_pred))
print("R2 Score:",metrics.r2_score(y_test,y_pred))
```

The Mean Absolute Error (MAE) is a metric used to evaluate the accuracy of a predictive model, typically regression models. It measures the average absolute difference between the predicted values and the actual values.
