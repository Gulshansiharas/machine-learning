```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
da=pd.read_csv('/content/drive/MyDrive/winequality-red.csv')
da
```

```python
da.head()
```

```python
da.tail()
```

```python
da.describe()
```

```python
da.isnull().sum()
```

```python
X = da[['fixed acidity', 'volatile acidity', 'alcohol']].values
y = da['quality'].values
X
```

```python
z=da[['pH','sulphates','alcohol']].values
z
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)
X_train,X_test,y_test
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(10)
df1
```

```python
df1.plot(kind='bar',figsize=(10,8))
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print("R2 Score: ", metrics.r2_score(y_test, y_pred))
```

```python
coefficients = regressor.coef_
intercept = regressor.intercept_
```

```python
print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
import seaborn as sns
# Plotting the linear regression plot
plt.figure(figsize=(10, 8))
sns.regplot(x='Actual', y='Predicted', data=df1 )  # Using seaborn for regression plot
plt.title('Linear Regression Plot: Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.grid(True)
plt.show()
```
